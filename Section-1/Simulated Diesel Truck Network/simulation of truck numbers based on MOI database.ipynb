# number of vehicles produced between the years

```python
import pandas as pd
import numpy as np



# Generate a range of years from the first to the last year in the DataFrame
all_years = range(vehicle_df['Year'].min(), vehicle_df['Year'].max() + 1)

# Interpolate the values for all years
interpolated_values = np.interp(
    all_years,
    vehicle_df['Year'],
    vehicle_df['Thousand Vehicles Produced']
)

# Create a new DataFrame with the interpolated values
interpolated_df = pd.DataFrame({
    'Year': all_years,
    'Thousand Vehicles Produced': interpolated_values
})

print(interpolated_df)
```

# number of vehicles produced in the key timeframes and their percentage over the total trucks

```python
# Define the ranges for each Euro standard
euro_ranges = {
    'euro3': (2001, 2004),
    'euro4': (2005, 2009),
    'euro5': (2010, 2013),
    'euro6': (2014, 2024)
}

# Initialize dictionaries to hold the sums and percentages
euro_totals = {}
euro_percentages = {}

# Calculate the total number of trucks produced
total_trucks = interpolated_df['Thousand Vehicles Produced'].sum()

# Calculate the sum of produced vehicles and percentage for each Euro standard range
for euro, (start_year, end_year) in euro_ranges.items():
    sum_produced = interpolated_df[
        (interpolated_df['Year'] > start_year - 1) & (interpolated_df['Year'] <= end_year)
    ]['Thousand Vehicles Produced'].sum()
    euro_totals[euro] = sum_produced
    percentage = (sum_produced / total_trucks) * 100
    euro_percentages[euro] = percentage

print("euro_total" ,euro_totals)

print("percentages" , euro_percentages)
```
